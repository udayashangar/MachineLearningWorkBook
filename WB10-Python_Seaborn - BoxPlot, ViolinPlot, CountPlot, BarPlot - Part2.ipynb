# Seaborn

## Categorical Plots


* boxplot
* violinplot
* countplot
* barplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### countplot

```python
df = sns.load_dataset('tips')
df.head()
```

```python
tip = df['sex'].value_counts()
tip
```

```python
sns.countplot('sex',data=df)
```

```python
day = df['day'].value_counts()
day
```

```python
sns.countplot('day',data=df)
```

```python
time = df['time'].value_counts()
time
```

```python
sns.countplot(y='time',data=df)
```

## BarPlot

```python
sns.barplot(x='sex',y='total_bill', data=df)
```

```python
sns.barplot(x='time',y='size', data =df)
```

## BoxPlot

```python
sns.boxplot(y='total_bill', x='size',data=df)
```

```python
sns.boxplot(data=df, orient='v')
```

```python
sns.boxplot(x='total_bill',y='day',hue='smoker',data=df)
```

## Violin Plor

```python
sns.violinplot(x='total_bill', y='day', data=df, palette='rainbow')
```

```python
iris = sns.load_dataset('iris')
iris
```

```python
sns.barplot(x='species',y='petal_length', data=iris)
```
