# Seaborn 

### Distribution Plot
#### - DistPlot
#### - JoinPlot
#### - PairPlot

```python
## importing seaborn lib
import seaborn as sns
```

```python
## loading the dataset 'tips'(which already exist in seaborn lib)
df=sns.load_dataset('tips')
```

```python
## displaying dataset
df.head()
```

```python
## find the datatypes of diff features in dataset
df.dtypes
```

# Correlation with HeatMap

```python
## find the correlation between dataset features
df.corr()
```

```python
## displaying the heatmap for the Correlation values
sns.heatmap(df.corr())
```

# Join Plot

### Univariate Analysis

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind="hex")
```

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind = "reg")
```

# Pair plot
### Can helps to do analysis with more than one features(mostly by pairing)

```python
## displaying pairplot for dataset
sns.pairplot(data=df)
```

```python
## displaying pairplot for dataset using hue
sns.pairplot(data=df,hue='sex')
```

```python
sns.pairplot(data=df, hue='smoker')
```

```python
df['smoker'].value_counts()
```

# Dist Plot

#### It helps to check the distribution of the column features

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False, bins=10)
```

```python
sns.distplot(df['tip'],kde=False)
```
