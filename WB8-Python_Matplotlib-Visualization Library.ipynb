# Matplotlib-Visualization Library

```python
## importing the matplotlib library
import matplotlib.pyplot as plt

## This one time declaration helps in replacing multiple <plot.show()> declarationsabs
%matplotlib inline
```

```python
## import Numpy
import numpy as np
```

```python
## assigning x & y axis values
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
## print the graph
plt.scatter(x,y,c='r')

## assigning x & y axis label
plt.xlabel('X axis')
plt.ylabel('Y axis')

## assigning Graph title
plt.title('2D Graph')

##print plot
##plt.show()

## Save the Graph
plt.savefig('test2.png')
```

```python
##Creating new values for y
y=x*x
```

```python
## using plot
plt.plot(x, y, color='green', marker='*', linestyle='dashed', linewidth=2, markersize=12)

## Assigning x and y axis
plt.xlabel('X axis')
plt.ylabel('Y axis')

## Graph Title
plt.title('2D Diagram')
```

```python
## using subplot

plt.subplot(2,3,2)
plt.plot(x,y,'r-')

plt.subplot(2,3,1)
plt.plot(x,y,'g*')

plt.subplot(2,3,4)
plt.plot(x,y,'bo')

plt.subplot(2,3,3)
plt.plot(x,y,'y')
```

```python
x = np.arange(0,15)
y = x*3+5

plt.plot(x,y,'y')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.title('Graph Diagram')
```

```python
## using sine and pi values

x=np.arange(0,np.pi*4,0.1)
y=np.sin(x)

plt.plot(x,y,'r--')
plt.xlabel('X axis')
plt.xlabel('Y axis')
plt.title('Sine Wave Graph')
```

```python
## using sine &cosine values
x = np.arange(0,5*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2,1,1)
plt.plot(x,y_sin,'r--')
plt.title('sine')

plt.subplot(2,1,2)
plt.plot(x,y_cos,'g')
plt.title('cossine')
```

```python
## Using Bar plots

x=[1,5,9]
y=[11,15,19]

x2=[2,4,7]
y2=[4,9,3]

plt.bar(x,y)
plt.bar(x2,y2,color='g')
plt.title('Bar Chart')
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

## Histogram

```python
a=np.array([2,3,4,15,34,54,67,32,7,66,56,83,33,45,21,26,53])
plt.hist(a, bins=20)
plt.title('Histogram')
```

# Box Plot using Matplotlib

```python
## using Box Plot

##Creating data using normal std
data =[np.random.normal(0,std,100) for std in range(1,4)]

plt.boxplot(data,vert='True',patch_artist ='b')
plt.show()
```

# Pie Chart

```python
## Plot

x=[222,199,160,100]
labels="Python","c++","JAVA","R Studio"
colors=['green','blue','red','grey']
explode=(0.5,0,0,0)

plt.pie(x, explode=explode, labels=labels, colors=colors, shadow=True, autopct='%1.1f%%')
         
plt.axis('equal')
plt.show()
```

```python
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.4, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=False)

plt.axis('equal')
plt.show()
```
